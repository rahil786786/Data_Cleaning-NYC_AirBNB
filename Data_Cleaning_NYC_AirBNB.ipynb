
# Data Columns Description

## Numerical Columns
- **id**: Unique identifier for each listing (integer).
- **host_id**: Unique identifier for the host (integer).
- **latitude**: Geographical latitude of the listing location (float).
- **longitude**: Geographical longitude of the listing location (float).
- **price**: Listing price per night (integer).
- **minimum_nights**: Minimum number of nights required to book (integer).
- **number_of_reviews**: Total number of reviews for the listing (integer).
- **reviews_per_month**: Average number of reviews per month (float).
- **calculated_host_listings_count**: Number of listings the host has (integer).
- **availability_365**: Number of days the listing is available in the next 365 days (integer).

## Categorical Columns
- **name**: Name/title of the listing (text/string).
- **host_name**: Name of the host (string).
- **neighbourhood_group**: Larger area/neighborhood grouping (e.g., borough) (string).
- **neighbourhood**: Specific neighborhood of the listing (string).
- **room_type**: Type of room offered (e.g., Entire home/apt, Private room) (string).
- **last_review**: Date of the last review (string/datetime).

## Mixed Columns
- Columns such as **last_review** can be treated as datetime for analysis but are strings in raw data.


# Univariate Analysis

```python
from scipy import stats
import numpy as np

# Select only numeric columns
numeric_df = df.select_dtypes(include=[np.number])

# Calculate Z-scores for each numeric column
z_scores = np.abs(stats.zscore(numeric_df))

# Keep only rows where all Z-scores < 3 (means not an outlier)
df_no_outliers = df[(z_scores < 3).all(axis=1)]

# Compare shapes to see how many rows were removed
print("Original shape:", df.shape)
print("After removing outliers:", df_no_outliers.shape)
```

# Bivariate Analysis Top 10 Hosts Name by Total Revenue

```python
# Top 10 hosts by total price (revenue)
top_hosts_by_price = df.groupby('host_name')['price'].sum().sort_values(ascending=False).head(10)

# Plot pie chart
plt.figure(figsize=(6, 6))
plt.pie(top_hosts_by_price, labels=top_hosts_by_price.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('viridis'), explode=[0.05]*10, shadow=True)
plt.title('Top 10 Hosts by Total Revenue')
plt.tight_layout()
plt.show()
```

# Bivariate Analysis Average Price per Room Type

```python
# Average Price per Room Type
avg_price_per_room_type = df.groupby('room_type')['price'].mean().sort_values(ascending=False)
ax = sns.barplot(x=avg_price_per_room_type.index, y=avg_price_per_room_type.values, hue=avg_price_per_room_type.index, edgecolor='black')
plt.title('Average Price by Room Type')
plt.ylabel('Average Price')
plt.xlabel('Room Type')
# Add labels to the bars
for container in ax.containers:
    ax.bar_label(container, fmt='%.2f', size=10)
plt.show()
```

# Insights from Exploratory Data Analysis

Based on the analysis performed, here are some key insights:

* **Price Distribution:** The majority of listings are priced at lower points, but there is a significant presence of high-priced outliers, indicating a skewed distribution.
* **Minimum Nights:** Similar to price, the minimum nights required for a booking is also skewed, with most listings requiring a short stay, but some having very high minimum night requirements.
* **Reviews:** Both the number of reviews and reviews per month are heavily skewed, suggesting that a small number of listings are very popular and receive a high volume of reviews, while most have few.
* **Host Activity:** A small number of hosts manage a large proportion of the listings, indicating the presence of "superhosts" or professional property managers.
* **Availability:** Listings tend to be either mostly available or mostly unavailable throughout the year, with fewer listings having moderate availability.
* **Geographical Distribution:** Listings are concentrated in specific areas within the neighbourhood groups, reflecting the popular tourist and residential areas of NYC. Manhattan and Brooklyn have the highest number of listings.
* **Room Type Distribution:** "Entire home/apt" and "Private room" are the most common room types, while "Shared room" is less frequent.
* **Price by Room Type:** "Entire home/apt" listings have a significantly higher average price compared to "Private room" and "Shared room" listings.
* **Room Type by Neighbourhood Group:** The distribution of room types varies across neighbourhood groups, with Manhattan having a higher proportion of "Entire home/apt" listings and Brooklyn having a more balanced mix of "Entire home/apt" and "Private room" listings.
* **Top Hosts by Revenue:** A few hosts generate a significant portion of the total revenue, aligning with the observation of hosts having a large number of listings.
